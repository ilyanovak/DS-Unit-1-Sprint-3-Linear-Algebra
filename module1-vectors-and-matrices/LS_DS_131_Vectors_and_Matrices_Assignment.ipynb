```
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 1.1 Create a two-dimensional vector and plot it on a graph

```
v1 = np.array([1,2])

fig, ax = plt.subplots()
ax.grid()
plt.xlim(0,3)
plt.ylim(0,3)
plt.arrow(0,0,
          *v1,
          head_width=.1,
          length_includes_head=True)
plt.show();
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
v2 = np.array([1,2,3])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([0,4])
ax.set_ylim([0,4])
ax.set_zlim([0,4])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.quiver(0,0,0,
          *v2)
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi

v1_5 = v1 * 5
v1_pi = v1 * pi
v1_neg_e = v1 * -1 * e

print(v1_5)
print(v1_pi)
print(v1_neg_e)

fig, ax = plt.subplots()
ax.grid()
plt.xlim(-7,7)
plt.ylim(-11,11)
plt.arrow(0,0,*v1_5,head_width=.5,linewidth=5,length_includes_head=True, color='red')
plt.arrow(0,0,*v1_pi,head_width=.5,length_includes_head=True, color='blue')
plt.arrow(0,0,*v1_neg_e,head_width=.5,length_includes_head=True, color='yellow')
plt.show();
```

They are all scaled versions of the initial vector

## 1.4 Graph vectors $\vec{a}$ and $\vec{b}$ and plot them on a graph

\begin{align}
\vec{a} = \begin{bmatrix} 5 \\ 7 \end{bmatrix}
\qquad
\vec{b} = \begin{bmatrix} 3 \\4 \end{bmatrix}
\end{align}

```
a = np.array([5,7])
b = np.array([3,4])

print(a)

fig, ax = plt.subplots()
ax.grid()
plt.xlim(0,6)
plt.ylim(0,8)
plt.arrow(0,0,*a,head_width=.5,length_includes_head=True, color='red')
plt.arrow(0,0,*b,head_width=.5,length_includes_head=True, color='blue')
plt.show();
```

## 1.5 find $\vec{a} - \vec{b}$ and plot the result on the same graph as $\vec{a}$ and $\vec{b}$. Is there a relationship between vectors $\vec{a} \thinspace, \vec{b} \thinspace \text{and} \thinspace \vec{a-b}$

```
a_minus_b = a-b

fig, ax = plt.subplots()
ax.grid()
plt.xlim(0,6)
plt.ylim(0,8)
plt.arrow(0,0,*a,head_width=.5,length_includes_head=True, color='red')
plt.arrow(0,0,*b,head_width=.5,length_includes_head=True, color='blue')
plt.arrow(0,0,*a_minus_b,head_width=.5,length_includes_head=True, color='yellow')

plt.show();
```

The x value of a-b is the difference between the x values of a and b; and likewise for its y value

##  1.7 Find $e \times f$

\begin{align}
\vec{e} = \begin{bmatrix} 5 \\ 7 \\ 2 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\4 \\ 6 \end{bmatrix}
\end{align}

```
ee = np.array([[5],[7],[2]])
f = np.array([[3],[4],[6]])
ee_cross_f = np.cross(ee,f, axis=0)
print(ee_cross_f)
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([[1],[1],[1],[8]])
h = np.array([[3],[3],[3],[3]])

g_norm = np.linalg.norm(g)
h_norm = np.linalg.norm(h)

print(g_norm)
print(h_norm)
```

g is longer than h
